```python
import pandas as pd, numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.datasets import boston_housing
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
```

```python
df = pd.read_csv('data/SampleSubmission2023.csv')
df
```

```python
pat1 = r'\d{4}_(\d{4})_\d{4}'
pat2 = r'\d{4}_\d{4}_(\d{4})'

idList = sorted(list(
    set(df['ID'].str.extract(pat1, expand=False)).union(set(df['ID'].str.extract(pat2, expand=False))) 
    ))
```

```python
menList = [int(i) for i in idList[:idList.index('3101')]]
womenList = [int(i) for i in idList[idList.index('3101'):]]
```

```python
len(menList)
```

```python
len(womenList)
```

```python
deprecated_mens = []
deprecated_womens = []

for i in range(menList[0], menList[-1] + 1):
    if i not in menList:
        deprecated_mens.append(i)

for i in range(womenList[0], womenList[-1] + 1):
    if i not in womenList:
        deprecated_womens.append(i)
```

```python
deprecated_womens
```

```python
m, w = len(menList), len(womenList)
```
